```python
import pandas as pd
filePath = 'Documents\DataFiles\AppleStore.csv'
def readCsvData(filePath):
    return pd.read_csv(filePath) 
f = readCsvData(filePath)
print(f)
```

```python
f.columns
```

```python
t=f[['prime_genre','user_rating']]
g=t[t.prime_genre=='Games']
ng=t[t.prime_genre!='Games']
print("Mean {0}".format(g.mean()))
print("mean not games {0}".format(ng.mean()))
```

```python
movies=['free app','Affordable apps','Expensive app','Very Expensive']
rating1=f[f.price==0]
rating2=f[(f.price>0)&(f.price<20)]
rating3=f[(f.price>20)&(f.price<50)]
rating4=f[f.price>50]
free=rating1.count()[0]
af_ap=rating2.count()[0]
ea=rating3.count()[0]
ve=rating4.count()[0]
li=[free,af_ap,ea,ve]
plt.pie(li,labels=movies,shadow=True,startangle=90,autopct="%1.1f%%")
plt.show()
```
